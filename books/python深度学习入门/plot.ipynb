```python
%%latex
$\Large

\begin{align}{
    a = b+c+d \\
    x = \frac{a}{y+z} = \frac{b+c+d}{y+z} \\
    z = \sum_{i}^{n} zz_{i}
}

\end{align}
$
```

```python
x = np.linspace(0, 2, 3)

y = np.linspace(0, 2, 3)

X, Y = np.meshgrid(x, y)
Z = np.zeros((3, 3))
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_aspect(1)
ax.plot_wireframe(X, Y, Z)
Z1 = np.ones((3, 3))
Z2 = Z1 * 2
ax.plot_wireframe(X, Y, Z1)
ax.plot_wireframe(X, Z, Y)
ax.plot_wireframe(Z, X, Y)
ax.plot_wireframe(X, Y, Z2)
ax.plot_wireframe(Z2, Y, X)
ax.plot_wireframe(Y, Z2, X)
ax.set_zlabel('f(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
