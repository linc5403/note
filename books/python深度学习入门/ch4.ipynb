## 4.2 损失函数

### 4.2.1 均方误差 (mean squared error)

```python
%%latex
\begin{equation}

E = \frac{1}{2}\sum_{k} (y_{k} - t_{k})^{2}

\end{equation}
```

```python
def mean_squared_error( y, t):
    return 0.5 * np.sum((y - t)** 2)
```

### 4.2.2 交叉熵误差（cross entropy error）

```python
%%latex
\[
    E = -\sum_{k} t_{k}\log y_{k}
\]
```

```python
x = np.arange(0.01, 1, 0.01)
y = -np.log(x)
plt.grid()
plt.xlabel('x')
plt.ylabel('-log(x)')
plt.plot(x, y)
```

**由上图可知，概率越趋近于1，则交叉熵误差越小；反之则误差越大**  

```python
def cross_entropy_error(y, t):
    delta = 1e-7
    return -np.sum(t * log(y + delta))
```

### 4.2.3 mini-batch学习

**所有训练数据的损失函数的总和作为训练的标准**  

```python
%%latex
\[
    E = - \frac{1}{N}\sum_{n}\sum_{k}t_{nk}\log y_{nk}
\]
```

```python
def func2(x, y):
    return x**2 + y**2
n = 20
size = 10
x = np.linspace(-size, size, n)
y = np.linspace(-size, size, n)

X, Y = np.meshgrid(x, y)
Z = func2(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_zlabel('f(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
